### Handling missing values

For categorical variables (host_location, host_neighbourhood, host_response_time) NA values will be replaced with "Missing". It is worth noting here, that there will be many missing values for host_neighbourhood (around half of the dataset)
For beds, the missing data will be imputed with the number of accomodates. 
Given that there are more than 27,000 observations in the dataset, other observations with missing data will be dropped. This will still mean a dataset of around 23,000 observations.

```python
data["host_location"] = data["host_location"].fillna("Missing")
data["host_neighbourhood"] = data["host_neighbourhood"].fillna("Missing")
data['beds']=data['beds'].fillna(data['accommodates'])

to_filter=data.isna().sum()
to_filter[to_filter>0]
```

```python
# for numerical columns, all missing data is to be dropped

data.dropna(subset=to_filter[to_filter>0].index, inplace = True)
```

## Modeling
For predictive modeling 3+1 models will be built. One with OLS by focusing on the accommodation related variables only and picking functional forms based on scatterplots (this model will be kept very simple intentionally as it will only serve as a reference for the other models), one with random forest using all variables, and a third one with Gradient Boosting. Then we will revert back with the most important variables of the ML algorithms and create an OLS with them. To compare results, we will look at cross-validated RMSE scores.

```python
from sklearn.linear_model import LinearRegression
```

```python
ols_vars = accom_vars + ["bathrooms_2", "minimum_nights_2"]
ols_vars
```

```python
# splitting the data to train and holdout in 80-20% ratio

data_train, data_holdout = train_test_split(data, train_size=0.8, random_state=42)
data_train.shape, data_holdout.shape
```

```python
# creating OLS model and checking RMSE value

y, X = dmatrices("price ~ " + " + ".join(ols_vars), data_train)

ols_model = LinearRegression().fit(X,y)

y_hat = ols_model.predict(X)

ols_rmse = mean_squared_error(y,y_hat,squared=False)
ols_rmse
```

Not surprisingly, this model that only contains a rather arbitrary list of variables, and functional forms, performs with a rather high RMSE.

```python
data_train.price.mean()
```

```python
ols_rmse/data_train.price.mean()
```

The very simplistic OLS model performs with an RMSE of 55.38 which is around 44% of the mean of the Y variable in the train dataset.

### Random forest

```python
# running the RF on all possible x variables

all_vars = accom_vars + host_vars + review_vars
y, X = dmatrices("price ~ " + " + ".join(all_vars), data_train)
```

```python
import math
print('The theoretical recommended number of variables: {:.2f}.'.format(math.sqrt(len(X.design_info.column_names))))
```

```python
rfr = RandomForestRegressor(random_state = 42)
```

```python
# creating tune_grid with more and less max features than the theoretically optimal (21) and a few options for minimum sample size for stopping

tune_grid = {"max_features": [19, 21, 23, 25], "min_samples_leaf": [5, 10, 15]}
```

```python
rf_random = GridSearchCV(
    rfr,
    tune_grid,
    cv=5,
    scoring="neg_root_mean_squared_error",
    verbose=3,
)
```

```python
%%time
rf_model = rf_random.fit(X, y.ravel())
```

```python
# getting best tuning setup

df_rf_model_cv_results = pd.DataFrame(rf_model.cv_results_)[[
    'param_max_features', 'param_min_samples_leaf', 'mean_test_score']]
df_rf_model_cv_results.columns = ['max features', 'min node size', 'RMSE']
df_rf_model_cv_results.pivot(
    index = 'max features', 
    columns = 'min node size', 
    values = 'RMSE').round(2)*-1
```

```python
rf_model.best_score_
```

```python
rf_model.best_estimator_
```

Best estimator has 25 features and 5 minimum observations in each node. 

The random forest performs better than the very simplistic OLS, but not by much.

#### Checking feature importances

```python
pd.DataFrame(
    rf_model.best_estimator_.feature_importances_, 
    X.design_info.column_names)
```

```python
df_var_imp = pd.DataFrame(
    rf_model.best_estimator_.feature_importances_, 
    X.design_info.column_names)\
    .reset_index()\
    .rename({"index": "variable", 0: "imp"}, axis=1)\
    .sort_values(by=["imp"], ascending=False)\
    .reset_index(drop = True)

df_var_imp['cumulative_imp'] = df_var_imp['imp'].cumsum()
```

```python
df_var_imp.style.format({
    'imp': lambda x: f'{x:,.1%}',
    'cumulative_imp': lambda x: f'{x:,.1%}'})
df_var_imp.head()
```

```python
# set cutoff for 2% as we are interested in only the "top" variables

cutoff = 0.02
```

```python
df_cutoff = df_var_imp[df_var_imp.imp > cutoff].sort_values(by = 'imp')
ggplot(df_cutoff, aes("reorder(variable, imp)", "imp")) + geom_col() + coord_flip() + theme_bw() + ylab("Importance") + xlab("Variables")
```

In the most important variables we find a mixture of accommodation and review related variables. There are only a handful of host related ones though. This somewhat confirms the initial intuition that the host is less important than the accommodation itself and the reviews of it.

Let's look at the grouped variable importances.

```python
categorical_encoder = OneHotEncoder(handle_unknown="ignore")

preprocessing = ColumnTransformer([
    ("cat", categorical_encoder, cat_var),
    ("num", "passthrough", num_var)])

rf_pipeline = Pipeline(
    [("preprocess", preprocessing), 
     ("regressor", rf_model.best_estimator_)] # put RF model to pipeline
)
```

```python
%%time
rf_pipeline.fit(data_train[all_vars],data_train.price)
```

```python
%%time
result = permutation_importance(
    rf_pipeline,
    data_train[all_vars],
    data_train.price,
    n_repeats=10,
    random_state=45,
)
```

```python
pd.DataFrame(
        result.importances_mean,
        data_train[all_vars].columns).head()
```

```python
df_grouped_var_imp = pd.DataFrame(
        result.importances_mean,
        data_train[all_vars].columns)\
    .sort_values(by = 0, ascending = False)\
    .reset_index()\
    .rename({'index': 'variable', 0: 'imp'}, axis = 1)
df_grouped_var_imp['cumulative_imp'] = df_grouped_var_imp.imp.cumsum()
```

```python
df_grouped_var_imp.style.format({
    'imp': lambda x: f'{x:,.1%}',
    'cumulative_imp': lambda x: f'{x:,.1%}'})
df_grouped_var_imp.head()
```

Let's visualise the ones that have a higher than 2% importance.

```python
# visualising top grouped variables

cutoff_grouped = 0.02
df_cutoff_gr = df_grouped_var_imp[df_grouped_var_imp.imp > cutoff_grouped].sort_values(by = 'imp')
ggplot(df_cutoff_gr, aes("reorder(variable, imp)", "imp")) + geom_col() + coord_flip() + theme_bw() + ylab("Importance") + xlab("Variables")
```

2 feature importances have appeared newly: neighbourhood and host neighbourhood. This suggest that although one specific neighbourhood does not have a high importance, in general, the location/neighbourhood of the AirBnB is important. This is also confirmed by the high importance of the review scores on location. It is worth noting that in host_neighbourhood there were many missing values.

```python
df_grouped_var_imp.cumulative_imp.iloc[6]
```

### GBM

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# creating Gradient Boosting with a few tuning options

gbm = GradientBoostingRegressor(learning_rate=0.1, min_samples_split=20, max_features = 10, n_estimators = 50
                               )

tune_grid = {"n_estimators": [200, 300], "max_depth": [5, 10]}

gbm_model_cv = GridSearchCV(
    gbm,
    tune_grid,
    cv=5,
    scoring="neg_root_mean_squared_error",
    verbose=10,
    n_jobs=-1
)
```

```python
# grouping categorical variables with OneHotEncoder

categorical_encoder = OneHotEncoder(handle_unknown="ignore")

preprocessing = ColumnTransformer([
    ("cat", categorical_encoder, cat_var),
    ("num", "passthrough", num_var)])

gbm_pipe = Pipeline(
    [("preprocess", preprocessing), 
     ("regressor", gbm_model_cv)], verbose=True # put best model to pipeline
)
```

```python
%%time
gbm_pipe.fit(data_train[all_vars],data_train.price)
```

```python
# checking best tuning parameters

gbm_model_cv.best_estimator_
```

```python
# checking RMSE value

gbm_rmse = gbm_model_cv.best_score_*-1
gbm_rmse
```

GBM achieves a lower RMSE than Random Forest and than OLS.

```python
# checking feature importances in the GBM model

result2 = permutation_importance(
    gbm_pipe,
    data_train[all_vars],
    data_train.price,
    n_repeats=10,
    random_state=45,
)
```

```python
pd.DataFrame(
        result2.importances_mean,
        data_train[all_vars].columns).head()
```

```python
df_grouped_var_imp_gbm = pd.DataFrame(
        result2.importances_mean,
        data_train[all_vars].columns)\
    .sort_values(by = 0, ascending = False)\
    .reset_index()\
    .rename({'index': 'variable', 0: 'imp'}, axis = 1)
df_grouped_var_imp_gbm['cumulative_imp'] = df_grouped_var_imp_gbm.imp.cumsum()
```

```python
df_grouped_var_imp_gbm.head()
```

```python
interact = ["beds:neighbourhood_cleansed", "beds:acoommodates", "beds:bathrooms"]
ols_pred = vars_ols2 + interact
```

```python
data_train_ols, data_holdout_ols = train_test_split(data_ols, train_size=0.8, random_state=42)
```

```python
from sklearn.model_selection import KFold
from statsmodels.tools.eval_measures import mse,rmse
k = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
for train_index, test_index in k.split(data_train_ols):
    print(train_index, '\n', '\n', test_index, '\n')
```

```python
### Cross validate OLS with combining sklearn k-fold cross validation and statsmodels ols formula


def cv_reg(formula, data, kfold, robustse=None):
    regression_list = []
    predicts_on_test = []
    rsquared = []
    rmse_list = []

    # Calculating OLS for each fold

    for train_index, test_index in k.split(data):
        # print("TRAIN:", train_index, "TEST:", test_index)
        data_train, data_test = data.iloc[train_index, :], data.iloc[test_index, :]
        if robustse is None:
            model = smf.ols(formula, data=data_train).fit()
        else:
            model = smf.ols(formula, data=data_train).fit(cov_type=robustse)
        regression_list += [model]
        predicts_on_test += [model.predict(data_test)]
        rsquared += [model.rsquared]
        rmse_list += [rmse(data_train[formula.split("~")[0]], model.predict())]

    return {
        "regressions": regression_list,
        "test_predict": predicts_on_test,
        "r2": rsquared,
        "rmse": rmse_list,
    }


def summarize_cv(cvlist, stat="rmse"):
    result = pd.DataFrame(
        {"Model" + str(x + 1): cvlist[x][stat] for x in range(len(cv_list))}
    )
    result["Resample"] = ["Fold" + str(x + 1) for x in range(len(cvlist[0]["rmse"]))]
    result = result.set_index("Resample")
    result = pd.concat([result, pd.DataFrame(result.mean(), columns=["Average"]).T])
    return result
```

```python
cv1 = cv_reg(
    "price~bathrooms + neighbourhood_cleansed + review_scores_location + accommodates + beds + reviews_per_month + number_of_reviews_l30d", data_train_ols, k, "HC0")
cv2 = cv_reg(
    "price~bathrooms + neighbourhood_cleansed + review_scores_location + accommodates + beds + reviews_per_month + number_of_reviews_l30d  + bathrooms_2 + beds_2 + accommodates_2+review_scores_location_2+reviews_per_month_2+number_of_reviews_l30d_2", data_train_ols, k, "HC0")
cv3 = cv_reg("price~bathrooms+neighbourhood_cleansed+review_scores_location+accommodates+beds+reviews_per_month+number_of_reviews_l30d+bathrooms_2+beds_2+accommodates_2+review_scores_location_2 + reviews_per_month_2+number_of_reviews_l30d_2+beds*neighbourhood_cleansed+beds*accommodates+beds*bathrooms", data_train_ols, k, "HC0")
```

```python
cv_list = [cv1, cv2, cv3]
summarize_cv(cv_list)
```

The best performing model is Model3 with all polinomials and interactions, although model 2 is also quite close. It still performs lower than the GBM but better than the initial basic OLS model and the RF models.

```python
print(f'Original OLS RMSE: {round(ols_rmse, 2)}')
print(f'Random forest RMSE: {round(rf_model.best_score_*-1, 2)}')
print(f'GBM RMSE: {round(gbm_rmse,2)}')
print(f'Refined OLS RMSE: {round(summarize_cv(cv_list).iloc[-1,-1], 2)}')
```
